```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/My SAE Studies/CST184_Weekly_Activities/Week_01/renewable_energy.csv")
df.head()


```

```python
high_renewable = df[df["%Renewable"] > 50]
print(high_renewable)

```

```python
high_set = set(high_renewable["Country"])
print("Countries with >50% renewable energy:", high_set)
print("Cardinality:", len(high_set))

```

```python
emissions_dict = dict(zip(df["Country"], df["CO2_Emissions"]))
print(emissions_dict)
print(emissions_dict["Brazil"])
```

```python
plt.bar(emissions_dict.keys(), emissions_dict.values())
plt.xticks(rotation=90)
plt.title("CO₂ Emissions per Capita by Country")
plt.ylabel("Tonnes")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df["Population"], df["%Renewable"])

for i in range(len(df)):
    plt.text(df["Population"][i], df["%Renewable"][i], df["Country"][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df["Population"], df["%Renewable"])

asia = df[df["Region"] =="Asia"]
print(asia)
print(asia.shape)

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```
